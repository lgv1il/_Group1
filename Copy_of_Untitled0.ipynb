```python
from google.colab import drive

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
file_path = "/Student_performance_data _.csv"
df = pd.read_csv(file_path)
print("Dataset Overview:")

print(df.info())  # Show data types and non-null values for each column

missing_values = df.isnull().sum() # Show the number of missing values per column
print("\n missing_values")
print(missing_values)

summary_statistics = df.describe() # Descriptive statistics of the dataset
print("\n summary_statistics")
print(summary_statistics)

fig, axes = plt.subplots(3, 3, figsize=(15, 12)) # Create a 3x3 plot layout
fig.suptitle("Distribution of Numeric and Categorical Variables")  # Add a main title

# Define numerical and categorical columns
numeric_columns = ["Age", "StudyTimeWeekly", "Absences", "GPA"] # Numerical columns
categorical_columns = ["Gender", "ParentalEducation", "Tutoring", "ParentalSupport", "GradeClass"] # Categorical columns

# Plot histograms for numerical variables
for i, col in enumerate(numeric_columns):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i // 3, i % 3]) # Plot distribution of each variable
    axes[i // 3, i % 3].set_title(f"Distribution of  {col}")


  # Plot categorical value distribution
for i, col in enumerate(categorical_columns):
    sns.countplot(x=df[col], ax=axes[(i + len(numeric_columns)) // 3, (i + len(numeric_columns)) % 3])
axes[(i + len(numeric_columns)) // 3, (i + len(numeric_columns)) % 3].set_title(f" The number of values for each  {col}")
# Adjust plot layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  #  Set the chart layout
plt.show()


```
