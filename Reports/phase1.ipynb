Project Goal:

---
The goal of this project is to analyze a dataset on student performance to explore the key factors affecting their academic results. Using clustering and classification techniques, we aim to predict students' performance levels and provide insights for personalized educational strategies.

---

Information About Dataset:


Number of objects: 35880

Number of columns: 15

Dataset link: https://www.kaggle.com/datasets/rabieelkharoua/students-performance-dataset

Class Lable: GradeClase (Score)

---

Attributes:
  

1. StudentID:
Each student is assigned a unique numerical identifier, ranging from 1001 to 3392, to distinguish them within the dataset.
 • Type: (Nominal)

 2. Age:
Represents the students’ ages in years, ranging from 15 to 18.
 • Type: Numeric (ratio-scaled)

 3. Gender:
Indicates the gender of each student, encoded numerically:
 • 0: Male
 • 1: Female

 • Type: Binary (symmetric)

 4. Ethnicity:
Describes the ethnicity of the students:
 • 0: Caucasian
 • 1: African American
 • 2: Asian
 • 3: Other

 • Type: Nominal.

 5. ParentalEducation:
The education level of the parents, This attribute assesses the impact of parental education on students’ academic success, coded as follows:
 • 0: None
 • 1: High School
 • 2: Some College
 • 3: Bachelor’s Degree
 • 4: Higher

 • Type: (Ordinal)

 6. StudyTimeWeekly:
Measures the total number of hours a student dedicates to studying per week, ranging from 0 to 20 hours.
 • Type: Numeric (ratio-scaled).

 7. Absences:
Tracks the number of school days a student has missed during the academic year, ranging from 0 to 30.
 • Type: Numeric (ratio-scaled).

 8. Tutoring:
Indicates whether a student receives private tutoring to enhance their learning:
 • 0: No
 • 1: Yes

 • Type: Binary (asymmetric).

 9. ParentalSupport:
Reflects the level of support provided by parents, coded as:
 • 0: None
 • 1: Low
 • 2: Moderate
 • 3: High
 • 4: Very High

 • Type: Ordinal.

 10. Extracurricular:
Denotes whether a student participates in extracurricular activities:
 • 0: No
 • 1: Yes

 • Type: Binary (asymmetric).

 11. Sports:
Indicates whether a student participates in sports activities:
 • 0: No
 • 1: Yes

 • Type: Binary (asymmetric).

 12. Music:
Captures whether a student engages in music-related activities:
 • 0: No
 • 1: Yes

 • Type: Binary (asymmetric).

 13. Volunteering:
Identifies students who engage in volunteering activities:
 • 0: No
 • 1: Yes

 • Type: Binary (asymmetric).

 14. GPA (Grade Point Average):
Represents the student’s cumulative academic performance on a scale from 2.0 to 4.0. GPA is influenced by study habits, parental involvement, and extracurricular activities.
 • Type: Numeric (interval-scaled).

 15. GradeClass :
Classifies students’ academic performance into grade categories based on their GPA:
 • 0: ‘A’ (GPA ≥ 3.5)
 • 1: ‘B’ (3.0 ≤ GPA < 3.5)
 • 2: ‘C’ (2.5 ≤ GPA < 3.0)
 • 3: ‘D’ (2.0 ≤ GPA < 2.5)
 • 4: ‘F’ (GPA < 2.0)

 • Type: Ordinal.
